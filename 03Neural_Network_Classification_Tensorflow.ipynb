#To visulize our model predition lets create a funcation plot_decision_boundry(), this funcation will:




*   Take in a trained model, feature (X) and label (Y)
*   Create a meshgrid of diffrent X values
*   Make predication across the meshgrid
*   plot the prediciation as well a line between zone(where each unique class falls)





```python
import numpy as np

def plot_decision_boundry(model,X,Y):
  """
  Plots the decisions boundry created by a model predicting on X.
  """
  #Define the axis boundry of the plot and create a meshgrid

  x_min, x_max = X[:,0].min() - 0.1 , X[:,0].max() + 0.1
  y_min, y_max = X[:,1].min() - 0.1 , X[:,1].max() + 0.1
  xx, yy = np.meshgrid(np.linspace(x_min, x_max,100),
                       np.linspace(y_min, y_max,100)
                       )

  #Create X value (we'r going to predicition to these)
  x_in = np.c_[xx.ravel(), yy.ravel()] #stack 2D array together

  # Remove the line below that expands dimensions
  # x_in = np.expand_dims(x_in, axis=1)

  #Make predicition
  y_pred = model.predict(x_in)

  # check for multiclass

  if len(y_pred[0]) > 1 :
    print("doing multiclass classficiation")
    #we have to reshape prediciation to get them ready for ploting
    y_pred = np.argmax(y_pred,axis = 1).reshape(xx.shape)
  else:
   print("doing binary classficiation")
   y_pred = np.round(y_pred).reshape(xx.shape)

  # Define contour levels
  levels = np.linspace(y_pred.min(), y_pred.max(), 100)


  # Plot the decision boundry
  plt.contourf(xx, yy, y_pred, levels=levels, cmap=plt.cm.RdYlBu, alpha = 0.7) # corrected yy instead of y_pred in plt.contourf
  plt.scatter(X[:, 0], X[:, 1], c=Y, s=40,cmap=plt.cm.RdYlBu) # added c=Y for coloring points
  plt.xlim(xx.min(),xx.max())
  plt.ylim(yy.min(),yy.max())
```

#Finding the Ideal Learning Rate


```python
#set randonseed
tf.random.set_seed(42)

model_13 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(10,activation="softmax"),
])

#compile
model_13.compile(
    loss = tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ["accuracy"]
)

#create ideal learning rate

mode_13_lr_schedular = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-3 * 10**(epoch/20))

#fit
mode_13_history = model_13.fit(train_data_norm,
                               train_labels,epochs=40,
                               validation_data = (test_data_norm,test_labels),
                               callbacks=[mode_13_lr_schedular])
```

```python
#create a confuesion matrix

import itertools
from sklearn.metrics import confusion_matrix


def make_confusion_matrix(y_true,y_pred,classes=None,figsize=(10,10),text_size=15,norm_const=1):

  #create the confusion matrix

  cm = confusion_matrix(y_true,y_pred)

  cm_norm = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis] #normilize our confusion matrix

  n_classes = cm.shape[0]


  #Lets preffiy
  fig,  ax = plt.subplots(figsize = figsize)

  #Create a matrix plot
  cax = ax.matshow(cm,cmap = plt.cm.Blues)
  fig.colorbar(cax)

  #set labels to be classes
  if classes:
    labels = classes
  else:
    labels = np.arange(cm.shape[0])



  #labels  the axes

  ax.set(
      title = "Confusion Matrix",
      xlabel = "Predicted Label",
      ylabel = "True Label",
      xticks = np.arange(n_classes),
      yticks = np.arange(n_classes),
      xticklabels = labels,
      yticklabels = labels

  )

  #set x-axis labels to bottom
  ax.xaxis.set_label_position("bottom")
  ax.xaxis.tick_bottom()

  #adjust label size
  ax.xaxis.label.set_size(text_size)
  ax.yaxis.label.set_size(text_size)
  ax.title.set_size(text_size)


  #set thredslod for diffrent color
  threshold = (cm.max() + cm.min()) / 2

  #plot the texr each cell
  for i , j in itertools.product(range(cm.shape[0]),range(cm.shape[1])):
    plt.text(j,i,f"{cm[i,j]} ({cm_norm[i,j]*100:.1f}%)",
             horizontalalignment = "center",
             color = "white" if cm[i,j] > threshold else "black",
             size = text_size
             )
```

```python
import random

def plot_random_image(model,images,true_labels,class_names):
  """
  Pick a random image, plot it and label it with prediction and truth label
  """
  #set up random integer
  i = random.randint(0,len(images))

  #create a predicition and tragets
  traget_image = images[i]
  pred_prod = model.predict(traget_image.reshape(1,28,28))
  pred_label = class_names[pred_prod.argmax()]
  true_label = class_names[true_labels[i]]

  #plot the image
  plt.imshow(traget_image,cmap=plt.cm.binary)

  #Change the color title depending on if the prediction right or wrong

  if pred_label == true_label:
     color = "green"

  else:
      color = "red"

  #Add xlabel information (prediction/true label)
  plt.xlabel("Pred:{}{:2.0f}% (True:{})".format(pred_label,100*tf.reduce_max(pred_prod),true_label),color=color)
```
