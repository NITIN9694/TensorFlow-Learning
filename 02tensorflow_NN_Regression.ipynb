###Setp in Modeling with tensorflow

1.Creating Model :- define the input and output layers, as well as the hidden layer of deeplearning model.

2. Compile Model:-
 define the loss funcation (in the words tell the models how wrong is it) and
 optimzer (tell our model how to imporve the pattern its learning)
 evaluation metrics (what we can use to interpret the performnce of our model)

3. Fiting Model:- letting the model try to find the patterns between X and y.(feature and labels)


````python
#set random seed
tf.random.set_seed(42)

#1.Creating Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

#2. Complile the modle
model.compile(loss = tf.keras.losses.mae,#mae is short for mean absolute error
              optimizer = tf.keras.optimizers.SGD (),  ## is short for stochastic gradient desecnt
              metrics = ["mae"]
              )
#3 fit the model
model.fit(tf.expand_dims(X,axis = -1),Y,epochs=5)
````

````python
X,Y
````

````python
#try and predict the model
model.predict(tf.constant([17.0]))
````

````python
#les't rebuild model
#1. create model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

#2. compile model
model.compile(
    loss= tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.SGD(),
    metrics = ["mae"]

)

#3 fit model

model.fit(tf.expand_dims(X,axis=-1),Y,epochs=50)
````

````python
model.predict(tf.constant([17.0]))
````

````python
#let's create the model (this time with extra hidden layer with 100 hidden uints)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(50,activation=None),
    tf.keras.layers.Dense(1)
])

#2 compile the model
model.compile(loss = "mae",
              optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
              metrics = ["mae"])
#3. fit the model

model.fit(tf.expand_dims(X,axis=-1),Y,epochs =100)
````

````python
#final got the answer
model.predict(tf.constant([17.0]))
````

##The 3 sets
**Training Set** - the model learn from data, which is typicly 70-80% of total data which is avilabel

**Validation Set** - the modele get tuned on this data , which is typcily 10-15% of avilabel data.

**Test Set** the model eveluated on this data,what is he learn which is typlicaly 10-15% of data

Ex:- Unversitry (trainingset) -> practice exam (validation set) -> final exam(test set)

these are called **Genralizition** :- The ability for a machin learning model to perfom well on data its hasn't seen

````python
 #split the data into train and test sets
 x_train = Vx[:40] #first 40 are training sample( 80% of data  )
 y_train = Yx[:40]

 x_test = Vx[40:] #last 10 are testing samples (20% of data)
 y_test = Yx[40:]


 len(x_train), len(y_train), len(x_test) , len(y_test)
````

````python
#creating the ploting funcation for comman use
def plot_predication(train_data = x_train,train_label = y_train, test_data = x_test ,test_label = y_test, predication = y_pred ):


  """
   Ploting training data, test data and compares predications to ground truth labels.

  """

  plt.figure(figsize=(10,7))
  #plot training data
  plt.scatter(train_data,train_label,c="b",label="Train Data")

  #plot test data
  plt.scatter(test_data,test_label,c="g",label="Test Data")

  #plot predicition
  plt.scatter(test_data,predication,c="r",label = "Predication")

  plt.legend()
````

````python
## A Large Example

import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
````

````python
#read in insureance data set
insurance = pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/refs/heads/master/insurance.csv")
insurance
````

````python
#let try onehot encode our dataset so ita all numbers
insurance_one_hot =  pd.get_dummies(insurance)
insurance_one_hot.head()
````

````python
#create x and y value (feature and labels)
X = insurance_one_hot.drop("charges",axis=1)
Y = insurance_one_hot["charges"]

#x view
X.head()


````

````python
#y view
Y.head()

````

````python
#creating training and test sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state = 42)

len(x_train),len(x_test),len(X),len(y_train)

````

````python
#build a neural network (sort of like model_2_above)

tf.random.set_seed(42)

insuranace_model = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

insuranace_model .compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.SGD(),
    metrics = ["mae"]
)

insuranace_model.fit(x_train,y_train,epochs=100)
````

````python
insuranace_model.evaluate(x_test,y_test)
````
